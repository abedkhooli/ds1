```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
from urllib.request import urlopen 
from collections import Counter
from datascience import *
import numpy as np
import re

def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())
```

```python
little_women_url = 'http://www.gutenberg.org/cache/epub/514/pg514.txt'
little_women_text = read_url(little_women_url)
chapters = little_women_text.split('CHAPTER ')[1:]
```

```python

Table().with_columns([
    "Jo",   np.char.count(chapters, "Jo"),
    "Meg",  np.char.count(chapters, "Meg"),
    "Amy",  np.char.count(chapters, "Amy"),
    "Beth", np.char.count(chapters, "Beth"),
    "Laurie", np.char.count(chapters, "Laurie")
]).cumsum().plot(overlay=True)
```

```python
Table().with_columns([
        "Characters", [len(c) for c in chapters],
        "Periods", np.char.count(chapters, "."),
    ]).scatter('Periods')
```

```python
t = Table().with_columns([
        "Characters", [len(c) for c in chapters],
        "Periods", np.char.count(chapters, "."),
        "Words", np.char.count(chapters, " "),
        "Word Length", [len(c) for c in chapters]/np.char.count(chapters, " "),
        "Sen. Length", [len(c) for c in chapters]/np.char.count(chapters, ".")
    ])
```

```python
t.show(50)
```

```python
t.hist('Sen. Length')
```

```python
t.sort('Sen. Length').show()
```

```python
# what about Arabic text? #2
ar_url3 = 'texts/hersh.txt' #saved as utf-8
with open(ar_url3, 'r') as content_file2:
    content2 = content_file2.read()
charz = len(content2)
wordz = np.char.count(content2, " ")
sentz = np.char.count(content2, ".")
print("total chars: ", charz)
print("words: ", wordz)
print("sentences: ", sentz)
print("Ave. word length: ", charz/wordz, " chars")
print("Ave. sent. length: ", charz/sentz, " chars")
print("Ave. words in sent.: ", wordz/sentz, " words" )
```
