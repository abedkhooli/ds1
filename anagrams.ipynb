```python
# read file and return clean word list
def read_word_list(filename):
    """Open file and create a dictionary of {"word":"sorted word"}."""
    with open(filename, "r") as word_file:
        word_list = {}
        for text in word_file:
            word = text.lower().strip()
            simple_text = "".join(sorted(word))
            word_list.update({word: simple_text})
        return word_list
    
#words = ['bad', 'app', 'sad', 'mad', 'dab','pge', 'bda', 'ppa', 'das', 'dba']
words = read_word_list("words")
anagrams = {}
for word in words:
    sword = ''.join(sorted(word))
    anagrams.setdefault(sword, []).append(word)

anagrams_list = [v for v in anagrams.values() if len(v) > 1]
print (anagrams_list)
```

```python
def printAnagrams():
    words = read_word_list("words")
    anagrams = {}
    for word in words:
        sword = ''.join(sorted(word))
        anagrams.setdefault(sword, []).append(word)

    anagrams_list = [v for v in anagrams.values() if len(v) > 1]
    print (anagrams_list)
```

```python
from collections import defaultdict

def printAnagrams2():
    d = defaultdict(list)
    for ele in read_word_list("words"):
        d[''.join(sorted(ele))].append(ele)

    print(filter(lambda x: len(x) > 1, d.values()))
```

```python
%timeit printAnagrams2()
```

```python
def return_anagrams(word_list):
    d = {}
    out = set()
    for word in word_list:
        s = ''.join(sorted(word))
        try:
            out.add(d[s])
            out.add(word)
        except:
            d[s] = word
    return out
```

```python
%timeit return_anagrams(read_word_list("words"))
```
